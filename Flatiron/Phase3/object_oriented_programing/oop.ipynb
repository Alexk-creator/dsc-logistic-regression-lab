```python
# creating a class #__init__
#behaviours- activate,deactivate, deposit, withdraw, transfer, check balance
class BankAccount:
    def __init__(self, acc_no, balance):#instance method
        self.acc_no = acc_no#feature/attribute
        self.balance = balance
    def deposit(self, amount):
        self.balance += amount
        return self.balance
    def withdraw(self, amount):
        if amount > self.balance:
            raise Exception("Insufficient balance")
        
        self.balance -= amount
        return self.balance
    def check_balance(self):
        return self.balance     
```

```python
# create an instance of a class/
# object 
my_acc = BankAccount("123456789", 1000)
```

```python
# accessing attributes
print(my_acc.acc_no)
print(my_acc.balance)
```

```python
# call object methods(functions that we create inside a class)part1
# behavior
print(my_acc.check_balance())
my_acc.deposit(500)
print(my_acc.check_balance())
my_acc.withdraw(200)
print(my_acc.check_balance())
```

```python
# calling the instance method2
print(my_acc.deposit(500))
print(my_acc.withdraw(200))
print(my_acc.check_balance())
```

## Inheritance 
- Allows a new class(child) to inherite attributes & methods from an existing
class(parent)
- Python supports multiple inheritans; where a child can inherit from multiple classes

```python
#creating a class animal
class Animal:
    def __init__(self, name):
        self.name = name
    def move(self):
        return f"{self.name} is moving"
    
```

```python
class Dog(Animal):
    def bark(self):
        return f"{self.name}  Woof! Woof!"
    
class bird(Animal):
    def fly(self):
        return f"{self.name} can fly"    
```

```python
# creating instance of the class dog and bird
my_dog = Dog("Rover")
print(my_dog.move())
print(my_dog.bark())
```

# creating machine learning pipeline using OOP

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np

# create a class for machine learning pipeline
class reg_model:
    def __init__(self, X, y, model):
        self.X = X
        self.y = y
        self.model = model
        self.scaler = StandardScaler()  # tool
    def split_data(self, test_size=0.2, random_state=7):
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(self.X, self.y, test_size=test_size, random_state=random_state)
    def preprocess(self):
        self.X_train_scaled = self.scaler.fit_transform(self.X_train)
    def train(self):
        self.model.fit(self.X_train_scaled, self.y_train)
    def predict(self):
        self.y_pred = self.model.predict(self.X_test)
        return self.y_pred
    def evaluate(self):
        predictions = self.model.predict(self.X_test)
        mse = mean_squared_error(self.y_test, predictions)
        return mse
```

```python
# sample data
df = pd.DataFrame({
    "feature1": [1, 2, 3, 4, 5],
    "feature2": [10, 20, 30, 40, 50],
    "target": [3, 6, 9, 12, 15]     
})
X = df.drop(columns=['target'])
y = df['target']
```

```python
# Regression
model = LinearRegression()
reg = reg_model(X, y, model)
reg.split_data()
reg.preprocess()
reg.train()
reg.predict()

```

```python
# classification
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
reg = reg_model(X, y, model)
reg.split_data()
reg.preprocess()
reg.train()
reg.predict()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Load the Ames housing dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
data = pd.read_csv(url)
data.head()
```

```python
data.isnull().sum()
```

```python
# Filling missing valuesdata
data['horsepower'].fillna(data['horsepower'].mean(), inplace=True)
```

```python
# Select features and target variable
X = data[['horsepower']]
y = data['mpg'] 
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
print(X_train.shape, X_test.shape) 
print(y_train.shape, y_test.shape)
```

```python
# Fitting the model
model = LinearRegression()
model.fit(X_train, y_train) 
```

```python
# Prediction
y_pred = model.predict(X_test)      
print(y_pred[:5])  # Print first 5 predictions
```

```python
print(y_test[:5])  
```

```python
# Checking the metrics
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
# Prediction should match y_train
y_train_pred = model.predict(X_train)
```

```python
# Cchecking underfitting and overfitting
mse_training = np.sqrt(mean_squared_error(y_train, y_train_pred))
mse_test = np.sqrt(mean_squared_error(y_test, y_pred))
print(mse_training)
print(mse_test)
```
